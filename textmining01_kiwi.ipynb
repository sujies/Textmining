### 2.3 동사, 형용사, 부사만의 워드 클라우드

```python
#동사, 형용사, 부사 뽑아내는 경우

def extract_pos(text):
    result = kiwi.tokenize(text)
    for token in result:
        if token.tag in ['VV', 'VA', 'MAG']:
            yield token.form
```

### TTR(Type Token Ratio) 조사




화면 캡처 2024-09-26 100630.png

날개

```python


# 전체 단어 리스트 생성 (공백 기준 분리)
tokens = content.split()

# 고유한 단어 리스트 생성
unique_tokens = set(tokens)

# TTR 계산
ttr = len(unique_tokens) / len(tokens) if len(tokens) > 0 else 0

# 결과 출력
print("날개")
print(f"전체 단어 수 (Tokens): {len(tokens)}")
print(f"고유한 단어 수 (Types): {len(unique_tokens)}")
print(f"Type-Token Ratio (TTR): {ttr:.4f}")
```

메밀꽃 필 무렵

```python

# 전체 단어 리스트 생성 (공백 기준 분리)
tokens = flower.split()

# 고유한 단어 리스트 생성
unique_tokens = set(tokens)

# TTR 계산
ttr = len(unique_tokens) / len(tokens) if len(tokens) > 0 else 0

# 결과 출력
print("메밀꽃 필 무렵")
print(f"전체 단어 수 (Tokens): {len(tokens)}")
print(f"고유한 단어 수 (Types): {len(unique_tokens)}")
print(f"Type-Token Ratio (TTR): {ttr:.4f}")
```

### 두 텍스트의 TTR차이점 조사하기

'메밀꽃 필 무렵' 텍스트의 TTR은 약 79.39%로, 매우 높은 어휘적 다양성을 보이고 있습니다. 이는 텍스트 내에서 사용된 단어의 상당수가 고유한 단어로 구성되어 있다는 것을 의미합니다.

1. 메밀꽃 필 무렵의 TTR 값이 가장 높음 (0.7939)

**어휘적 다양성:**

* 메밀꽃 필 무렵 > 독도는 우리땅 > 날개 순으로 어휘적 다양성이 높습니다.
* 메밀꽃 필 무렵은 서정적이거나 묘사적 표현이 많아, 다양한 어휘를 사용한 반면, 날개는 반복적인 어휘 사용이 많아 단조로울 수 있습니다.

**텍스트 특성:**
* 메밀꽃 필 무렵은 문학적이고 풍부한 어휘를 사용하는 서술 방식으로 독자를 다양한 어휘로 이끕니다.
* 날개는 특정 주제나 사건을 반복적으로 다루며, 주제가 집중되고 어휘의 폭이 제한적일 수 있습니다.
* 독도는 우리땅은 짧은 텍스트에도 불구하고 비교적 다양한 어휘를 사용하여, 강한 메시지를 전달하는 효과가 있을 수 있습니다.
