품사 개수(날개)

```python
import re
from konlpy.tag import Kkma

# 특수문자 제거 함수
def clean_text(text):
    return re.sub(r'[^\x00-\x7F가-힣]+', ' ', text)

file_path = '/content/drive/MyDrive/텍스트마이닝/소설데이터/날개.txt'

# Kkma 형태소 분석기 사용
kkma = Kkma()

# 파일 읽기 및 특수 문자 제거 후 형태소 분석
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()
    cleaned_content = clean_text(content)


kkma.pos(cleaned_content)
kk1 = kkma.pos(cleaned_content)
```

```python
import pandas as pd
df1 = pd.DataFrame(kk1)
df1[1].value_counts()
```

### TTR(Type Token Ratio) 조사




화면 캡처 2024-09-26 100630.png

날개

```python
# 형태소 분석기 초기화
okt = Okt()

# 전체 단어 리스트 생성 (공백 기준 분리)
tokens = content.split()

# 고유한 단어 리스트 생성
unique_tokens = set(tokens)

# TTR 계산
ttr = len(unique_tokens) / len(tokens) if len(tokens) > 0 else 0

# 결과 출력
print("날개")
print(f"전체 단어 수 (Tokens): {len(tokens)}")
print(f"고유한 단어 수 (Types): {len(unique_tokens)}")
print(f"Type-Token Ratio (TTR): {ttr:.4f}")
```

### 두 텍스트의 TTR차이점 조사하기

1) '날개' 데이터

    전체 단어 수 (Tokens): 5,889
    고유 단어 수 (Types): 3,028
    Type-Token Ratio (TTR): 0.5142

'날개' 텍스트는 전체 단어의 약 51.42%가 고유한 단어입니다. 이는 비교적 높은 TTR 값으로, 텍스트 내에서 어휘적 다양성이 적절하게 나타나고 있음을 의미합니다.

    전체 단어 수 (Tokens): 2,067
    고유 단어 수 (Types): 1,641
    Type-Token Ratio (TTR): 0.7939

'메밀꽃 필 무렵' 텍스트의 TTR은 약 79.39%로, 매우 높은 어휘적 다양성을 보이고 있습니다. 이는 텍스트 내에서 사용된 단어의 상당수가 고유한 단어로 구성되어 있다는 것을 의미합니다.

3) 독도는 우리땅 텍스트

    전체 단어 수 (Tokens): 104
    고유한 단어 수 (Types): 66
    Type-Token Ratio (TTR): 0.6346

TTR 값이 0.6346로, 전체 단어의 약 63%가 고유한 단어입니다. 비교적 짧은 텍스트임에도 불구하고 어휘적 다양성이 중간 이상으로 나타나고, 특정 주제에 대해 다양한 표현을 사용한 것을 알 수 있습니다.

    * 특정 주제에 대한 다양한 관점을 제시하거나, 다양한 표현을 사용했음을 나타낼 수 있습니다.

3. 날개의 TTR 값이 가장 낮음 (0.5142)

    * 전체 단어 수는 많지만, 고유한 단어의 비율이 상대적으로 낮습니다.
    * 동일한 단어나 표현이 많이 반복되며, 특정 주제나 사건을 반복적으로 다루고 있는 텍스트일 가능성이 큽니다.


**어휘적 다양성:**

* 메밀꽃 필 무렵 > 독도는 우리땅 > 날개 순으로 어휘적 다양성이 높습니다.
* 메밀꽃 필 무렵은 서정적이거나 묘사적 표현이 많아, 다양한 어휘를 사용한 반면, 날개는 반복적인 어휘 사용이 많아 단조로울 수 있습니다.

**텍스트 특성:**
* 메밀꽃 필 무렵은 문학적이고 풍부한 어휘를 사용하는 서술 방식으로 독자를 다양한 어휘로 이끕니다.
* 날개는 특정 주제나 사건을 반복적으로 다루며, 주제가 집중되고 어휘의 폭이 제한적일 수 있습니다.
* 독도는 우리땅은 짧은 텍스트에도 불구하고 비교적 다양한 어휘를 사용하여, 강한 메시지를 전달하는 효과가 있을 수 있습니다.
